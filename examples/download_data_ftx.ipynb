### Downloading historical data

```python
start_date = dt.datetime(2020, 1, 1)
end_date = dt.datetime(2021, 8, 1)
resolution = 300  # 5 minutes candles 
market = 'ABNB/USD'
all_data = pd.DataFrame()

while start_date < dt.datetime(2021, 2, 1):
    result = m.get_historical_prices(market, resolution, start_date, start_date + dt.timedelta(days=6))
    start_date += dt.timedelta(days=6)
    all_data = all_data.append(pd.DataFrame(result['result']))    
```

```python
all_data.to_csv('data/ABNBUSD.csv', index=None)
```
